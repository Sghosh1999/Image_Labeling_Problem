```python
from fastai.vision import *
from fastai.metrics import accuracy
```

Image Augmentation

```python
tfms = get_transforms()
```

```python
path = "../input/dataset/"
```

```python
np.random.seed(0)
src = (ImageList.from_csv(path, 'train.csv', folder='Train Images')
      .split_by_rand_pct(0.2)
      .label_from_df())
```

```python
data = (src.transform(tfms, size=224).databunch().normalize(imagenet_stats))
```

```python
data
```

```python
get_ipython().run_line_magic('pinfo', '')
```

```python
data.show_batch(rows=3, figsize=(7,11))
```

```python
arch = models.resnet152
```

```python
from keras.utils.generic_utils import get_custom_objects
from keras import applications
from keras.layers import Activation
import keras
import tensorflow as tf
def swish(x,beta = 1):
    return x*sigmoid(beta*x)

get_custom_objects().update({'swish':Activation(swish)})
```

```python
#opt_func:Callable='Adam'
learn = cnn_learner(data, arch, metrics=accuracy)
learn.model_dir='/kaggle/working/'

```

```python
learn.lr_find(num_it=200)
learn.recorder.plot()
```

```python
#Fitting One Cycle Method in this Problem
lr = 1e-4
learn.fit_one_cycle(15,max_lr=lr)
```

```python
learn.save('autotagimagesstage-1')
```

```python
#Fitting One Cycle Method in this Problem
lr = 1e-4
learn.fit_one_cycle(5,max_lr=lr)
```

```python
learn.save('autotagimagesstage-1')
```

```python
# all layers now trainable
learn.unfreeze()
# optionally, separate LR and WD for each group
learn.lr_find()
learn.recorder.plot()
```
